```python
df['trans_date_trans_time']=pd.to_datetime(df['trans_date_trans_time'])
```

```python
df['day']=df['trans_date_trans_time'].dt.day
```

```python
df['hour'] = df['trans_date_trans_time'].dt.hour
```

```python
df['month']=df['trans_date_trans_time'].dt.month
```

```python
df['day_name']=df['trans_date_trans_time'].dt.day_name()
```

```python
df.head(5)
```

```python
df['high_amount']=df['amt']>df['amt'].quantile(0.95)
```

```python
df['is_night']=df['hour'].between(0,5)
```

```python
df.columns
```

```python
df[['Hour', 'hour']].head()
```

```python
df = df.drop(['Hour'], axis=1)
```

```python
df.columns
```

```python
df['amount_bucket']=pd.cut(df['amt'],
                          bins=[0,50,200,500,1000,5000,100000],
                          labels=['Very Low','Low','Medium','High','Very High','Extreme'])
```

```python
df['fraud_status']=df['is_fraud'].map({0:'Normal',1:'fraud'})
```
